```python
import pandas as pd
data = {"Courses": ["Math", "Biology", "Math", "Biology"],
        "Days": ["Saturday","Saturday","Sunday","Sunday"],
        "MinDuration":[2,4,4,2],
        "MaxDuration":[4,6,6,4]} 
df = pd.DataFrame(data)
df
```

```python
df_melt=pd.melt(df,
                id_vars=["Courses","Days"],
                value_vars=["MinDuration","MaxDuration"],
                var_name="DurationType",
                value_name="Duration")
df_melt
```

```python
df_pivot=df_melt.pivot(index=["Courses","DurationType"],
                       columns="Days",
                       values=["Duration"])
df_pivot
```

```python
df_pivot.columns = df_pivot.columns.droplevel()
df_pivot
```

```python
modified_df=df_pivot.rename_axis(None,axis=1)
modified_df
```

```python
modified_df=modified_df.reset_index()
modified_df
```

```python
df.transpose()
```

```python
df_melt.transpose()
```
